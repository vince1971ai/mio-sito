```python
# Importare le classi necessarie
from sentence_transformers import SentenceTransformer
import numpy as np # La useremo per gestire gli array di embeddings

# Assicurati che df_emotions sia definito dalla cella precedente.
# (Lo sarà se hai eseguito la cella che mi hai appena mostrato)

# Caricare un modello pre-addestrato
print("Caricamento del modello Sentence Transformer...")
model_st = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
print("Modello caricato.")

# Ottenere la lista dei nomi delle emozioni dal DataFrame
emotion_names = df_emotions['emotion'].tolist()
print(f"\nNomi delle emozioni da cui generare gli embeddings: {emotion_names[:5]}...") # Mostra i primi 5

# Generare gli embeddings per i nomi delle emozioni
print("\nGenerazione degli embeddings in corso... (potrebbe richiedere un po' di tempo)")
emotion_embeddings = model_st.encode(emotion_names, show_progress_bar=True)
print("Embeddings generati.")

# Aggiungere gli embeddings come nuova colonna nel DataFrame
df_emotions['embeddings'] = emotion_embeddings.tolist()

# Visualizzare le prime righe del DataFrame con la nuova colonna embeddings
print("\nDataFrame con la nuova colonna 'embeddings':")
print(df_emotions.head())

# Controlliamo la forma di un embedding per capire la sua dimensione
print(f"\nDimensione del primo embedding (numero di features): {len(df_emotions['embeddings'].iloc[0])}")
print(f"Tipo di dato nella colonna embeddings: {type(df_emotions['embeddings'].iloc[0])}")
```

```python
# Assicurati che df_emotions sia definito dalle celle precedenti
# e che contenga la colonna 'embeddings'.

# 1. Prendiamo gli "embeddings" come INPUT (X) per il modello
# La colonna 'embeddings' contiene liste di numeri.
# Dobbiamo trasformarla in un formato che i modelli di machine learning capiscono meglio
# (una specie di tabella di numeri, dove ogni riga è un embedding).
X = pd.DataFrame(df_emotions['embeddings'].to_list())

# 2. Prendiamo le coordinate x, y, z come OUTPUT (y) che il modello dovrà imparare a predire
y = df_emotions[['x', 'y', 'z']]

# 3. Mostriamo le prime righe dei nostri Input (X) e Output (y) per vedere se è tutto a posto
print("Prime righe degli INPUT (X) per il modello (gli embeddings):")
print(X.head())

print("\nPrime righe degli OUTPUT (y) che il modello dovrà predire (le coordinate x, y, z):")
print(y.head())

print(f"\nForma degli Input (X): {X.shape}")
print(f"Forma degli Output (y): {y.shape}")
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split # Anche se non lo useremo subito per dividere, è buona prassi importarlo
from sklearn.preprocessing import StandardScaler # Per normalizzare i dati
import numpy as np # Già importato, ma per chiarezza

# Assicurati che X e y siano i DataFrame/Array che abbiamo preparato prima
# X dovrebbe avere forma (41, 384) e y (41, 3)

# 0. Preparazione degli Input (Normalizzazione)
# Le reti neurali spesso funzionano meglio se i dati di input sono "normalizzati"
# (cioè riscalati in modo che abbiano media 0 e deviazione standard 1).
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X) # Adatta e trasforma X

# Anche se l'output y (coordinate) non sempre necessita di scaling per la regressione,
# in alcuni casi può aiutare la convergenza del modello, specialmente se gli intervalli sono molto diversi.
# Per ora, proviamo a non scalarlo o a scalarlo con cautela se necessario,
# ma teniamo presente che è un'opzione. Per questo primo giro, usiamo y così com'è.

print("Input X scalati (prime 5 righe):")
print(X_scaled[:5])
print(f"Forma di X_scaled: {X_scaled.shape}")

# 1. Creare il Modello MLPRegressor
# Spiegazione semplice dei parametri:
# - hidden_layer_sizes=(100, 50): Significa che la nostra rete avrà 2 "strati nascosti" di neuroni.
#   Il primo strato avrà 100 neuroni, il secondo ne avrà 50. Questi neuroni fanno i calcoli.
# - max_iter=1000: Il modello proverà ad aggiustarsi fino a 1000 volte (epoche) sui dati.
# - activation='relu': È un tipo di "funzione di attivazione" che i neuroni usano per decidere se "sparare" un segnale. 'relu' è molto comune.
# - solver='adam': È l'algoritmo che il modello usa per imparare e aggiustare i suoi pesi. 'adam' è un buon tuttofare.
# - random_state=42: Questo assicura che se rieseguiamo il codice, otteniamo gli stessi risultati "casuali", rendendo l'esperimento ripetibile.
# - learning_rate_init=0.001: Quanto velocemente il modello impara.

print("\nCreazione del modello MLPRegressor...")
mlp_model = MLPRegressor(hidden_layer_sizes=(128, 64, 32), # Proviamo una rete un po' più profonda
                         activation='relu',
                         solver='adam',
                         max_iter=2000, # Aumentiamo un po' le iterazioni
                         random_state=42,
                         learning_rate_init=0.001,
                         early_stopping=True, # Fermiamoci se non migliora più
                         n_iter_no_change=50, # Numero di iterazioni senza miglioramento prima di fermarsi
                         verbose=True) # Mostra i messaggi di addestramento
print("Modello creato.")

# 2. Addestrare il Modello
# Diamo al modello gli embeddings scalati (X_scaled) e le coordinate originali (y)
print("\nInizio addestramento del modello...")
mlp_model.fit(X_scaled, y)
print("Addestramento completato.")

# 3. Ottenere le Predizioni sui Dati di Addestramento
# Chiediamo al modello di predire le coordinate usando gli stessi input su cui ha imparato.
print("\nOttenimento delle predizioni sui dati di addestramento...")
y_pred_train = mlp_model.predict(X_scaled)
print("Predizioni ottenute.")

# 4. Visualizzare un Confronto tra Reale e Predetto (per i primi esempi)
# Creiamo un piccolo DataFrame per confrontare facilmente.
comparison_df = pd.DataFrame({
    'Emozione': df_emotions['emotion'][:10], # Prendiamo solo le prime 10 per brevità
    'X_Reale': y['x'][:10],
    'Y_Reale': y['y'][:10],
    'Z_Reale': y['z'][:10],
    'X_Predetto': np.round(y_pred_train[:10, 0], 2), # Arrotondiamo a 2 decimali
    'Y_Predetto': np.round(y_pred_train[:10, 1], 2),
    'Z_Predetto': np.round(y_pred_train[:10, 2], 2)
})

print("\nConfronto tra Coordinate Reali e Predette (prime 10 emozioni):")
print(comparison_df)

# Possiamo anche calcolare un "punteggio" di quanto bene ha imparato sui dati di addestramento.
# Un valore di 1.0 è perfetto (cosa che potrebbe succedere con pochi dati).
train_score = mlp_model.score(X_scaled, y)
print(f"\nPunteggio del modello sui dati di addestramento (R^2 score): {train_score:.4f}")
print("(Un punteggio più vicino a 1.0 significa che il modello ha imparato bene i dati di addestramento)")
```

```python
import pandas as pd
import numpy as np

# Assicurati che df_new_emotions (dal CSV) sia ancora caricato e corretto.
# Se per caso il kernel si è riavviato o df_new_emotions si è perso,
# dovrai rieseguire la cella che lo carica da Google Drive.

if 'df_new_emotions' in locals() or 'df_new_emotions' in globals():
    print("Partendo da df_new_emotions (dati originali dal CSV)...")
    # 1. Rimuoviamo la prima riga (intestazioni) e creiamo una NUOVA copia
    df_frasi_emozioni_tentativo = df_new_emotions.iloc[1:].copy().reset_index(drop=True)

    colonne_da_forzare_numeric = ['id', 'valore1', 'valore2']

    print("\nTentativo di conversione FORZATA e controllo tipi:")
    for colonna in colonne_da_forzare_numeric:
        print(f"--- Processando colonna: {colonna} ---")
        # Applichiamo .str.strip() solo se la colonna è di tipo object e contiene stringhe
        if df_frasi_emozioni_tentativo[colonna].dtype == 'object':
            try:
                df_frasi_emozioni_tentativo[colonna] = pd.to_numeric(df_frasi_emozioni_tentativo[colonna].str.strip(), errors='coerce')
            except AttributeError: # Corretto: Aggiunto ':' qui
                df_frasi_emozioni_tentativo[colonna] = pd.to_numeric(df_frasi_emozioni_tentativo[colonna], errors='coerce')
        else:
            df_frasi_emozioni_tentativo[colonna] = pd.to_numeric(df_frasi_emozioni_tentativo[colonna], errors='coerce')

        if df_frasi_emozioni_tentativo[colonna].isnull().sum() == 0:
            try:
                df_frasi_emozioni_tentativo[colonna] = df_frasi_emozioni_tentativo[colonna].astype(np.int64)
                print(f"Colonna '{colonna}' convertita in int64.")
            except ValueError:
                print(f"Colonna '{colonna}' non può essere int64 (forse contiene decimali o stringhe non convertibili), lasciata come float64 se possibile.")
        else:
            print(f"Colonna '{colonna}' contiene NaN dopo pd.to_numeric. Sarà di tipo float64 se i restanti sono numeri.")

        print(f"Tipo di dato finale per '{colonna}': {df_frasi_emozioni_tentativo[colonna].dtype}")

    print("\n--- Informazioni finali sul DataFrame 'df_frasi_emozioni_tentativo' ---")
    df_frasi_emozioni_tentativo.info()

    print("\n--- Conteggio finale dei NaN ---")
    print(df_frasi_emozioni_tentativo.isnull().sum())

    if pd.api.types.is_numeric_dtype(df_frasi_emozioni_tentativo['id']) and \
       pd.api.types.is_numeric_dtype(df_frasi_emozioni_tentativo['valore1']) and \
       pd.api.types.is_numeric_dtype(df_frasi_emozioni_tentativo['valore2']):
        df_frasi_emozioni = df_frasi_emozioni_tentativo # Assegnazione finale
        print("\nDataFrame 'df_frasi_emozioni' aggiornato con successo con tipi numerici.")
        print(df_frasi_emozioni.head())
    else:
        print("\nATTENZIONE FINALE: Le colonne numeriche non sono state tutte convertite correttamente.")

else:
    print("ERRORE CRITICO: Il DataFrame 'df_new_emotions' (dal CSV) non è definito.")
    print("Per favore, esegui prima la cella che carica il file CSV da Google Drive.")
```
